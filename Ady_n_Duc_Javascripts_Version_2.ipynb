# 3.0 Share-data dictionary and App class


- Start with a "Shareable" global variable (a dictionary) where you can pass data between the native Python and the NodeJS/Javascript.

- The reason for sharable data is we will use NodeJS/Javascript for the logic and the "widgets" is from "ipywidgets" package. 

><h2>The "ShareData" can be accessing from both NodeJS and Python.</h2>

```
%%node
global.ShareData = {"version": "0.2.0"};
```

- As a good little programmer, we start by creating an object or class in Javascripts.

```
%%node
class App {
  constructor() {
    ShareData["css"] = {"txt": ' style="padding: 1rem; font-size:1.3rem;font-family:cursive;" '};
  }
  //
  print(msg) {
    html('<div ' + ShareData["css"]["txt"] + '>' + msg + '</div>');
  }
}
```

```
%%node
kid = new App()
```

# 4.0 Create a "Card" Style

```
%%node
// test it
kid.print("Hello Ady");
```

- You can update the css by update the shareable object (dictionary), and add new method as needed.

```
%%node
ShareData["card_bg"] = '#20c997';
ShareData["css"]["card"] = ' style="padding: 2rem; border: 1px solid #333344;border-radius: 5px;color: #ffffff;font-family: serif;font-size:5rem;text-align: center; display: inline-block;box-shadow: 3px 3px 8px rgba(0,0,0,0.2);background-color: ';
App.prototype.printCard = function (msg) {
  html('<div ' + ShareData["css"]["card"] + ShareData["card_bg"] + ';">' + msg + '</div>');
};
```

```
%%node
kid.printCard("a");
```

```
%%node
kid.printCard("B");
```

# 5.0 Create [Python] Slider and the value in NodeJS.

- Using ipyWidgets
https://ipywidgets.readthedocs.io/en/stable/examples/Widget%20Basics.html

```
# Python
import ipywidgets 
from IPython.display import display
```

```
# Python
age = ipywidgets.widgets.IntSlider(
    value=7,
    min=1,
    max=20,
    description='Age:'
)
```

```
# Python
display(age)
```

```
# Python
# Save it on Sharedata nd check the value
ShareData["age"] = age.value
print(ShareData["age"])
```

- Access the "age" in NodeJS via "ShareData" variable.

```
%%node
// check it in NodeJS/Javascript
kid.printCard(ShareData["age"]);
```

# 6.0 [Python] Color Picker to change NodeJS's Card Background Color

```
#Python
card_bg = ipywidgets.widgets.ColorPicker(
    concise=False,
    description='Pick Background Color: ',
    value='pink',
    disabled=False
)
display(card_bg)
```

```
# Python
# Save it on Sharedata nd check the value
ShareData["card_bg"] = card_bg.value
print(ShareData["card_bg"])
```

```
%%node
// check it in NodeJS/Javascript
kid.printCard("Hello Ady");
```

# 7.0 [Python] Event

```
# Python
words = ipywidgets.widgets.ToggleButtons(
  options=['Apple', 'Banana', 'Carrot', "Date-fruit"],
  description='Alphabets:',
  disabled=False,
  button_style='warning', # 'success', 'info', 'warning', 'danger' or ''
  icons=['check']
)
#
def on_words_clicked(evt):
  ShareData["word"] = words.value
  return
#
words.observe(on_words_clicked)
#
display(words)
```

```
%%node
kid.printCard(ShareData["word"])
```
